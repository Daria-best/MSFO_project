Для обучающей выборки возьмем 70 процентов от датасета по каждому банку, упорядоченному по времени

```python
def train_test(data):
    train_index=[]
    for i in data['cregnum'].unique():
        indexes=data.loc[data['cregnum']==i].index
        train_index.extend(indexes[0:round(len(indexes)*(1-0.3))])
    train=data_class.loc[data.index.isin(train_index)]
    test=data_class.loc[~data.index.isin(train_index)]
    x = train.iloc[:,2:-1]
    y = train.iloc[:,-1:]
    x_test=test.iloc[:,2:-1]
    y_test = test.iloc[:,-1:]
    return x, y, x_test, y_test, train_index  
```

```python
x, y, x_test, y_test, train_index=train_test(data_class)
```

Меры - recall демонстрирует способность алгоритма обнаруживать данный класс вообще, а precision — способность отличать этот класс от других классов

#### Функция, которая будет выводить оценки качества моделей

```python
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import auc
from sklearn.metrics import f1_score

def metrics_(test, pred):
    report = classification_report(test, pred, target_names=['0', '1'])
    print(report)
    print('f1_score: %.3f' % f1_score(test, pred))
    # Матрица правдоподобия
    class_names = [0, 1]
    fig, ax = plt.subplots()
    ticks = np.arange(len(class_names))
    plt.xticks(ticks, class_names, size=15)
    plt.yticks(ticks, class_names, size=15)
    conf_matr=metrics.confusion_matrix(test, pred)
    conf_matr_norm=(conf_matr/conf_matr.sum(axis=1))
    #print(conf_matr)
    #print(conf_matr.sum(axis=1))
    
    sns.set()
    sns.heatmap(pd.DataFrame(conf_matr_norm),
                annot=True, cmap='summer', annot_kws={"size": 15}) 
    plt.ylabel('Действительные значения', size=15)
    plt.xlabel('Предсказанные значения', size=15)
    plt.title('\nНормализованная матрица путаницы\n', fontsize=15)
```
